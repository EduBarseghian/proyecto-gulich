```python
longitoo=-61.807861
latitoo=-31.528142

fechasagua=[]
```

```python
listadefechas = []
anios=range(2010,2020)
fechasporanio=range(1,365,8)
for yy in anios:
    for d in fechasporanio:
        listadefechas.append(str(yy)+"%03d"%d) 
```

```python
listaindices=["ndwi","mndwi"]
dicci={listaindices[i]:i for i in range(len(listaindices))}

listaceeseves=[]
for k in range(3):  
    listaceeseves.append([])
    for indice in listaindices:
        listaceeseves[k].append([])
```

```python
#coso= gdal.Open(directoriotifs+"ndwi_2010129_500m.tif") #MANU
coso= gdal.Open(directoriotifs+"ndwi/ndwi_2010129_500m.tif")  #EDU
transform = coso.GetGeoTransform()
#xOrigin = transform[0], yOrigin = transform[3], pixelWidth = transform[1], pixelHeight = -transform[5]
a=transform[0]
b=transform[1]
c=transform[3]
d=transform[5]
del coso
```

```python
for i in range(len(listadefechas)): 
    for indice in listaindices:
        in_ds = gdal.Open(directoriotifs+indice+"/"+indice+"_"+listadefechas[i] + "_500m.tif")  
        print(indice+"_"+listadefechas[i])
        in_band = in_ds.GetRasterBand(1)
        in_data = in_band.ReadAsArray(0, 0, in_ds.RasterXSize, in_ds.RasterYSize)

        for k in range(3):                   
            col = int((longitoo - a) / b)
            row = int((latitoo - c) / d)                
            arreglo = []
            for l in range(-k,k+1):
                for j in range(-k,k+1):
                    arreglo.append(in_data[row+l][col+j])
            listaceeseves[k][dicci[indice]].append(max(-2.0,np.mean(np.array(arreglo))))
    del in_ds
    del in_band            
    del in_data            
```

```python
datos=[]
for indice in listaindices:
    datos.append([])
for k in range(3):
    for indice in listaindices:
        datos[listaindices.index(indice)] = pd.DataFrame(data = listaceeseves[k][dicci[indice]])
        datos[listaindices.index(indice)] = datos[listaindices.index(indice)].rename(columns = {0:"promedio"})
        datos[listaindices.index(indice)]['indice'] = [indice]*len(listadefechas)
        datos[listaindices.index(indice)]['fecha'] = listadefechas
```

```python
num = []
label=[]
for i in range(len(listadefechas)):
    if i%2 == 0:
        num.append(i)
        label.append(listadefechas[i])
```

```python
result = pd.concat(datos[dicci[indice]] for indice in listaindices)
        
plt.figure(figsize=(70,6))
sns.lineplot(data=result, x='fecha', y = 'promedio', hue='indice')
plt.ylabel('promedio')
plt.xlabel('año y fecha')
plt.title("indices por fecha en el área alrededor del centrito")
plt.title("Promedio índices. Centro en ("+str(longitoo)+" , "+str(latitoo)+")")
plt.xticks(rotation=90)
plt.xticks(num, label)

xpoints = a=[len(fechasporanio)*i for i in range(1,len(anios))]

for p in xpoints:
    plt.axvline(p,  color = 'black')

plt.axhline(0,color = 'pink')
plt.axhline(-0.2,color = 'pink')
plt.axhline(-0.4,color = 'pink')
plt.axhline(-0.6,color = 'pink')       
for fff in fechasagua:
    plt.axvline(julianomodis(fff), color = 'red')
```
