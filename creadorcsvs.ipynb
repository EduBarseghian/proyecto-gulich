```python
# PARA CREAR LOS CSV DE 1 PUNTITO, VIA SU LONGITUD Y LATITUD
# Que queden como lista
longitoo=[-61.807861]
latitoo=[-31.528142]

directoriosalida=directoriosancor
prefijonombres="puntito_"

rangodefilas=range(1)
```

```python
# VERA JUAN PABLO (correr solamente 1 de las 4 celdas)
df = pd.read_csv(directoriopuntos+"AguaVera21022019_trans.csv")
directoriosalida=directoriocsvs+"Vera_csvs/"
rangodefilas=range(54)          #estas son las filas cuyos csv queremos crear
prefijonombres="Vera_"

totaldefilas=54        #cantidaddefilas de df_puntos_agua
longitoo=[]
latitoo=[]  
for r in range(totaldefilas):
    longitoo.append(df.Long_trans[r])
    latitoo.append(df.Lat_trans[r])
```

```python
# DIEGO (correr solamente 1 de las 4 celdas)
df = pd.read_csv(directoriopuntos+"agua_2016_diego_100.csv")
directoriosalida=directoriocsvs+"diego_csvs/"
rangodefilas=range(100)          #estas son las filas cuyos csv queremos crear
prefijonombres="diego_"

totaldefilas=100        #cantidaddefilas de df_puntos_agua
longitoo=[]
latitoo=[]
for r in range(totaldefilas):
    longitoo.append(float(df.Long_trans[r].replace(",",".")))
    latitoo.append(float(df.Lat_trans[r].replace(",",".")))
```

```python
# NICOLAS (correr solamente 1 de las 4 celdas)
df = pd.read_csv(directoriopuntos+"09_marzo_2015.csv")
directoriosalida=directoriocsvs+"nicolas_csvs/"
rangodefilas=range(29)          #estas son las filas cuyos csv queremos crear
prefijonombres="nicolas_"

totaldefilas=29
longitoo=[]
latitoo=[]
for r in range(totaldefilas):
    longitoo.append(float(df.Longitud[r].replace(",",".")))
    latitoo.append(float(df.Latitud[r].replace(",",".")))
```

```python
# NICOLAS FILTRADO(correr solamente 1 de las 4 celdas)
df = pd.read_csv(directoriopuntos+"aguanopermanente_filtrada.csv")
directoriosalida=directoriocsvs+"nico_agua_no_permanente_csvs/"
rangodefilas=range(29)          #estas son las filas cuyos csv queremos crear
prefijonombres="aguanoperm_"

totaldefilas=29
longitoo=[]
latitoo=[]
for r in range(totaldefilas):
    longitoo.append(float(df.Longitud[r].replace(",",".")))
    latitoo.append(float(df.Latitud[r].replace(",",".")))
```

```python
listaindices=["ndwi","mndwi"]
dicci={listaindices[i]:i for i in range(len(listaindices))}

listaceeseves=[]
for k in range(3):  
    listaceeseves.append([])
    for e in rangodefilas:
        listaceeseves[k].append([])
        for indice in listaindices:
            listaceeseves[k][e].append([])
```

```python
coso= gdal.Open(directoriotifs+"ndwi_2010129_500m.tif")
transform = coso.GetGeoTransform()
#xOrigin = transform[0], yOrigin = transform[3], pixelWidth = transform[1], pixelHeight = -transform[5]
a=transform[0]
b=transform[1]
c=transform[3]
d=transform[5]
del coso
```

```python
for i in range(len(listadefechas)): 
    for indice in listaindices:
        in_ds = gdal.Open(directoriotifs+indice+"/"+indice+"_"+listadefechas[i] + "_500m.tif")  
        print(indice+"_"+listadefechas[i])
        in_band = in_ds.GetRasterBand(1)
        in_data = in_band.ReadAsArray(0, 0, in_ds.RasterXSize, in_ds.RasterYSize)

        for k in range(3):  
            for e in rangodefilas:               
                col = int((longitoo[e] - a) / b)
                row = int((latitoo[e] - c) / d)                
                arreglo = []
                for l in range(-k,k+1):
                    for j in range(-k,k+1):
                        arreglo.append(in_data[row+l][col+j])
                listaceeseves[k][e][dicci[indice]].append(max(-2.0,np.mean(np.array(arreglo))))
    del in_ds
    del in_band            
    del in_data            
```

```python
for e in rangodefilas:
    for k in range(3):
        for indice in listaindices:
            datos = pd.DataFrame(data = listaceeseves[k][e][dicci[indice]])
            datos=datos
            datos = datos.rename(columns = {0:"promedio"})
            datos['indice'] = [indice]*len(listadefechas)
            datos['fecha'] = listadefechas
            nombresalida=prefijonombres+ str(e).zfill(3)+"_"+indice+"_"+str(2*k+1)+"x"+str(2*k+1) 
            pd.DataFrame(datos).to_csv(directoriosalida+nombresalida + ".csv")
            del datos
```
